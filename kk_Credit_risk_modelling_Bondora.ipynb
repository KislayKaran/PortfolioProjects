### **Outlier Treatment to remove outliers from Numerical Features**

```python
features_with_outliers = ["VerificationType","MonthlyPayment","IncomeTotal","DebtToIncome","FreeCash","PlannedInterestTillDate","CurrentDebtDaysPrimary",
                          "CurrentDebtDaysSecondary","PrincipalOverdueBySchedule","PlannedPrincipalPostDefault","PlannedInterestPostDefault","EAD1","EAD2","PrincipalRecovery",
                          "InterestRecovery","RecoveryStage","CreditScoreEeMini","PrincipalWriteOffs","PreviousRepaymentsBeforeLoan","PreviousEarlyRepaymentsBefoleLoan",
                          "NrOfScheduledPayments","PrincipalDebtServicingCost","InterestAndPenaltyDebtServicingCost","PrincipalPaymentsMade","InterestAndPenaltyPaymentsMade",
                          "PrincipalWriteOffs","InterestAndPenaltyWriteOffs","PrincipalBalance","InterestAndPenaltyBalance","AmountOfPreviousLoansBeforeLoan",
                          "PreviousRepaymentsBeforeLoan","PreviousEarlyRepaymentsBefoleLoan"]
def outliers(df,feature):
  q1 = df[feature].quantile(0.25)
  q3 = df[feature].quantile(0.75)
  IQR = q3-q1
  lower_bound = q1 - (IQR*1.5)
  upper_bound = q3 + (IQR*1.5)
  outliers_list=df.index[ (df[feature] < lower_bound) | (df[feature] > upper_bound)]
  return outliers_list


  
```

```python
index_list=[]
for feature in features_with_outliers:
  index_list.extend(outliers(df, feature))

index_list  
```

```python
def remove(df,outliers_list):
  outliers_list=sorted(set(outliers_list))
  df=df.drop(outliers_list)
  return df
  
df= remove(df, index_list)
print(df.shape)
```

### **Imputing the column with the missing values**

```python
#Impute missing values in numerical features using mean
num_features_with_null = [feature for feature in num_features if df[feature].isnull().sum()]
for feature in num_features_with_null:
   mean_value = df[feature].mean()
   df[feature].fillna(mean_value,inplace=True)
```

```python
# Imputing the missing values in categorical features using the most frequent value which is mode
catg_features_with_null = [feature for feature in catg_features if df[feature].isnull().sum()]
for each_feature in catg_features_with_null:
  mode_val = df[each_feature].mode()[0]
  df[each_feature].fillna(mode_val,inplace=True)
```

## Classifying Status column into Binary **classificatio**

```python
# 0 for repaid Loan, 1 for late loan, 2 for Current loan
# create a list of our conditions
conditions = [(df['Status']=="Repaid"),(df['Status'] == "Late") ,(df['Status']=="Current")]


# create a list of the values we want to assign for each condition
values =[0,1,2]
# create a new column and use np.select to assign values to it using our lists as arguments
df['Status_count_loans'] = np.select(conditions, values)
 # display updated DataFrame
df.head()
```

```python

df["Status_count_loans"].dtype

df.loc[(df['Status_count_loans']>=0) & (df['Status_count_loans']<=1), 'Customer_loans'] = 0
df.loc[df['Status_count_loans']>1 , 'Customer_loans'] = 1

 ### As we classify data, 0 represent Repaid and late loans, 1 represent Current Loans
```
